```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
```

```python
#head정보 얻어오는곳: http://www.useragentstring.com/
#python으로 접근 막는 사이트를 우회하여 접근하도록
head = {
    'Host':'marketdata.krx.co.kr',
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36',
    'Referer': 'http://marketdata.krx.co.kr/mdi',
}

#otp 가져오기
param = {
    'bld': 'COM/finder_stkisu',
    'name': 'form'
}
otp_url = 'http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx?bld=COM%2Ffinder_stkisu&name=form&_=1599881700567'
ret = requests.get(otp_url,headers=head, params=param)
```

```python
print(ret.text)
```

```python
url = 'http://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx'
dat = {
    'no': 'P1',
    'mktsel': 'ALL', 
    'pagePath': '/contents/COM/FinderStkIsu.jsp',
    'pageFirstCall': 'Y',
}
dat['code'] = ret.text
total_df = pd.DataFrame()
for i in range(1,16):
    dat['consonant'] = i
    ret2 = requests.post(url,data = dat, headers = head)
    rt = json.loads(ret2.text)#json.loads는 json형식의 string을 python dic으로 변경해줌
    total_df = pd.concat([total_df,pd.DataFrame(rt['block1'])])
    
#dat['consonant'] = 1
#ret2 = requests.post(url,data=dat, headers=head)
```
