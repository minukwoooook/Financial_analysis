# 시가총액 데이터 다운로드

```python
driver.get('https://marketdata.krx.co.kr/mdi#document=040402')
```

```python
head = {
    'host':'marketdata.krx.co.kr',
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.18362',
}

param = {
    'bld': 'MKD/04/0404/04040200/mkd04040200_01',
    'name': 'form'
}
otp = requests.get('https://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx',headers=head, params = param).text
```

```python
otp
```

```python
dat = {
    'market_gubun': 'ALL',
    'sect_tp_cd': 'ALL',
    'schdate': '20200316',
    'pagePath': '/contents/MKD/04/0404/04040200/MKD04040200.jsp',
    'curPage': '1',
}
dat['code'] = otp
ret = requests.post('https://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx',data=dat,headers=head)
```

```python
ret.text
```

```python
rt_list=[]
for i in range(1,300):
    if(i%10 == 0):
        print(i)
    dat['curPage'] = i
    ret = requests.post('https://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx',data=dat,headers=head)
    if(len(ret.text)>100):
        rt_list.append(ret.text)
    else:
        break
```

```python
rt_list[0]
```

```python
json.loads(rt_list[0])['시가총액 상하위']
```

```python
total_df = pd.DataFrame()
pd.concat([total_df,pd.DataFrame(json.loads(rt_list[0])['시가총액 상하위'])])
```

```python
total_df = pd.DataFrame()
for x in rt_list:
    rt = json.loads(x)
    #print(rt['시가총액 상하위'])
    total_df = pd.concat([total_df,pd.DataFrame(rt['시가총액 상하위'])])
```

```python
total_df
```

```python
total_df.reset_index(inplace=True,drop=True)
```

```python
total_df['date'] = '20200316'
```
